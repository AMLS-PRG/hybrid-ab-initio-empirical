```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('progress.out', sep='\\s+', skip_blank_lines=True, skiprows=[1])
print(df.shape)
df.head()
```

```python
iteration = np.array(df['iteration'])
loss_energy = np.array(df['le_avg'])
loss_force = np.array(df['lf_avg'])
```

```python
plt.figure(figsize=(6, 4))
plt.plot(iteration, loss_energy*1000, label='RMSE energy (meV/atom)', color='red')
plt.plot(iteration, loss_force*1000, label='RMSE force (meV/)', color='orange')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.xlim(iteration[0], iteration[-1])
plt.yscale('log')
plt.title('Model Optimization')
plt.legend()
plt.grid(which='both', linestyle='--', linewidth=.75)
plt.tick_params(axis='y', which='both', right=True, labelright=True)  # <-- Aade esto
plt.show()
```
